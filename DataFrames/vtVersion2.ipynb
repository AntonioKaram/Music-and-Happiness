```python
import pandas as pd
import plotly.express as px
pd.set_option('display.max_rows', None)

store = pd.HDFStore('store.h5')
df = store['df']
print(df)

import plotly.express as px
fig = px.choropleth(df, locations='country', color='depression', hover_name='country', animation_frame='year', range_color=[10, 128])
fig.show()
```
